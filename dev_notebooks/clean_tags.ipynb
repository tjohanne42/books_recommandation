```python
import pandas as pd
```

```python
df_t = pd.read_csv("databases/tags.csv")
df_bt = pd.read_csv('databases/book_tags.csv')
df_bt = df_bt.merge(df_t, on="tag_id")
```

## drop duplicates

```python
df_TEMP = df_bt.groupby(["goodreads_book_id", "tag_id"], as_index=False).count()
for book_id, tag_id in zip(df_TEMP.loc[df_TEMP["count"] == 2]["goodreads_book_id"], df_TEMP.loc[df_TEMP["count"] == 2]["tag_id"]):
    display(df_bt.loc[(df_bt["goodreads_book_id"] == book_id) & (df_bt["tag_id"] == tag_id)])
df_bt.drop_duplicates(subset=["goodreads_book_id", "tag_id"], inplace=True)
```

# create new df

```python
def create_df_with_tag_count(df):
    dictionnary = {"goodreads_book_id": [], "tag_id": [], "count": [], "tag_name": []}

    goodreads_book_id_list = df["goodreads_book_id"].unique()

    for goodreads_book_id in goodreads_book_id_list:

        df_book_tmp = df[df["goodreads_book_id"] == goodreads_book_id]
        df_book_tmp = df_book_tmp.reset_index(drop=True)

        dictionnary["goodreads_book_id"].append(goodreads_book_id)

        tag_id_list = []
        count_list = []
        tag_name_list = []

        for i in range(df_book_tmp.shape[0]):
            tag_id_list.append(df_book_tmp["tag_id"][i])
            count_list.append(df_book_tmp["count"][i])
            tag_name_list.append(df_book_tmp["tag_name"][i])

        dictionnary["tag_id"].append(tag_id_list)
        dictionnary["count"].append(count_list)
        dictionnary["tag_name"].append(tag_name_list)

    df_bt2 = pd.DataFrame(data=dictionnary)

    df_bt2["tag_count"] = df_bt2["tag_id"].map(lambda x: len(x))
    print(df_bt2["tag_count"].value_counts())
    
    df_bt2 = df_bt2.sort_values(by="tag_count", ascending=True)
    df_bt2 = df_bt2.reset_index(drop=True)

    display(df_bt2)
    return df_bt2
```

```python
my_list = df_bt["tag_name"].value_counts()
my_list = my_list.loc[(my_list < 3000) & (my_list > 1500)].index.tolist()
```

```python
dictionnary = {"tag_name": my_list}
df_tmp = pd.DataFrame(data=dictionnary)
```

```python
df_tmp2 = pd.merge(left=df_bt, right=df_tmp, on="tag_name", how="inner")
print(df_tmp2["tag_name"].unique().shape)
df_tmp2
```

```python
df_tmp3 = create_df_with_tag_count(df_tmp2)
```

```python
df_tmp2.to_csv("databases/book_tags2.csv")
```
